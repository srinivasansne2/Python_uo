```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img=cv.imread(r"C:\Bir.jpg",0)

plt.imshow(img,cmap="gray")
plt.show()

plt.hist(img.ravel(),256,[0,256])

plt.title("Histogram of Image")
plt.xlabel("No Of Pixels")
plt.ylabel("Gray Scale Pixels")
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img=cv.imread(r"C:\Bir.jpg")
src=cv.cvtColor(img,cv.COLOR_BGR2RGB)

color=("r","g","b")
for channel,col in enumerate(color):
    hist=cv.calcHist([src],[channel],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,256])
    plt.title("Histogram of Color Image")
    plt.xlabel("No Of Pixels")
    plt.ylabel("Gray Scale Pixels")
    
plt.show()
plt.imshow(src)
plt.show()
```
