```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"c:\Try.png")
```

```python
RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
print("original Dimansions:",RGB.shape)
```

```python
half=cv2.resize(RGB,(0,0),fx=0.1,fy=0.1)
bigger=cv2.resize(RGB,(1050,1610))
sketch_near=cv2.resize(RGB,(780,540),interpolation=cv2.INTER_NEAREST)
```

```python
Titel=["Original","Half","Bigger","Interpolation Nearest"]
images=[RGB,half,bigger,sketch_near]
count=4
print(type(Titel))
```

```python
for i in range(count):
    plt.subplot(2,2,i+1)
    plt.title(Titel[i])
    plt.imshow(images[i])
    
plt.show()
```

# Color

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
```

```python
img=cv2.imread(r"c:\Tes.jpg")
RGB=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
width=600
height=320
dim=(width,height)

resized=cv2.resize(RGB,dim,interpolation=cv2.INTER_AREA)
```

```python
grey=cv2.cvtColor(resized,cv2.COLOR_BGR2GRAY)
(thresh,BW)=cv2.threshold(grey,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(resized)
```

```python
plt.imshow(grey)
```

```python
plt.imshow(BW)
```

# Identifying Color

```python
img=cv2.imread(r"c:\Tes.jpg")
```

```python
HSV=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
UB=np.array([0,50,50])
LB=np.array([10,255,255])
mask=cv2.inRange(HSV,UB,LB)
```

```python
cv2.imshow("Original",img)
cv2.imshow("Mask",mask)
```
